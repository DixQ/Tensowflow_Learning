```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
physical_devices = tf.config.list_physical_devices('GPU') 
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
from tensorflow import keras
```

# Data

```python
data = pd.read_csv('IPN31152N.csv', parse_dates = True, index_col = 'DATE')
```

```python
data.info()
```

```python
data.head()
```

```python
data.plot(figsize = (12,8))
```

```python
test_size = 25
test_ind = len(data) - test_size
```

```python
train = data.iloc[:test_ind]
test  = data.iloc[test_ind:]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train)
```

```python
scaled_train = scaler.transform(train)
```

```python
scaled_test  = scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 12
generator = TimeseriesGenerator(scaled_train, scaled_train, length = length, batch_size = 1)
```

# Model

```python
n_features = 1
```

```python
model = keras.models.Sequential([
    keras.layers.LSTM(50, activation = 'relu', input_shape = (length, n_features), return_sequences="True"),
    keras.layers.LSTM(100, activation = 'relu'),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', loss = 'mse')
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', patience = 2)
```

```python
validation_generator = TimeseriesGenerator(scaled_test, scaled_test, length=length, batch_size = 1)
```

```python
model.fit_generator(generator, epochs = 20, validation_data = validation_generator, callbacks = [early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
test_predictions = []

first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(test)):
    
    current_pred = model.predict(current_batch)[0]
    
    test_predictions.append(current_pred)
    
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis = 1)
```

```python
true_pred = scaler.inverse_transform(test_predictions)
```

```python
test['Pred'] = true_pred
```

```python
test.plot()
```
